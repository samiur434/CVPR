```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("CP.csv")
df.head()
```

```python
X = df[['X1', 'X2']].values
Y = df['Class'].values


Y_onehot = np.zeros((Y.size, 3))
Y_onehot[np.arange(Y.size), Y - 1] = 1


X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
class NeuralNetwork(object):
    def __init__(self):
        self.inputLayerNeurons = 2
        self.hidden1Neurons = 10
        self.hidden2Neurons = 8
        self.outputNeurons = 3

        self.learning_rate = 0.1


        self.W_H1 = np.random.randn(self.inputLayerNeurons, self.hidden1Neurons)
        self.W_H2 = np.random.randn(self.hidden1Neurons, self.hidden2Neurons)
        self.W_O = np.random.randn(self.hidden2Neurons, self.outputNeurons)

        plt.show()

    def sigmoid(self, x, der=False):
        if der:
            return x * (1 - x)
        return 1 / (1 + np.exp(-x))

    def feedForward(self, X):

        self.hidden_input1 = np.dot(X, self.W_H1)
        self.hidden_output1 = self.sigmoid(self.hidden_input1)


        self.hidden_input2 = np.dot(self.hidden_output1, self.W_H2)
        self.hidden_output2 = self.sigmoid(self.hidden_input2)


        self.output_input = np.dot(self.hidden_output2, self.W_O)
        self.pred = self.sigmoid(self.output_input)
        return self.pred
    def backPropagation(self, X, Y, pred):

        output_error = Y - pred
        output_delta = output_error * self.sigmoid(pred, der=True)


        hidden2_error = output_delta.dot(self.W_O.T)
        hidden2_delta = hidden2_error * self.sigmoid(self.hidden_output2, der=True)

        hidden1_error = hidden2_delta.dot(self.W_H2.T)
        hidden1_delta = hidden1_error * self.sigmoid(self.hidden_output1, der=True)
        self.W_O += self.hidden_output2.T.dot(output_delta) * self.learning_rate
        self.W_H2 += self.hidden_output1.T.dot(hidden2_delta) * self.learning_rate
        self.W_H1 += X.T.dot(hidden1_delta) * self.learning_rate
    def train(self, X, Y):
        err = []
        for i in range(10000):
            output = self.feedForward(X)
            self.backPropagation(X, Y, output)
            loss = np.mean(np.square(Y - output))
            err.append(loss)
        plt.plot(err)
        plt.title("Training Loss")
        plt.xlabel("Epoch")
        plt.ylabel("MSE")
        plt.show()

    def predict(self, X):
        pred = self.feedForward(X)
        return np.argmax(pred, axis=1) + 1
```

```python
NN = NeuralNetwork()
```

```python
err = []
for i in range(10000):
    NN.train(X, Y_onehot)
    loss = np.mean(np.square(Y_onehot - NN.feedForward(X)))
    err.append(loss)
```
